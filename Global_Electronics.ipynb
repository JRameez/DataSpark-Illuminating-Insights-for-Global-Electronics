### CONVERTING TO NECESSARY DATATYPES

#### Customer - Birthday

```python
#Convert Birthday to Datetime format

customer['Birthday']=pd.to_datetime(customer['Birthday'], format='%m/%d/%Y')
#df['Birthday'] = pd.to_datetime(df['Birthday'], format="%Y-%m-%d").dt.strftime('%d-%m-%Y')
```

```python
customer.info()
```

```python
## Calcualting the age from birthday
from datetime import datetime

def calculate_age(birthday):
    today = datetime.now()
    age = today.year - birthday.year - ((today.month, today.day) < (birthday.month, birthday.day))
    return age

customer_copy = customer
customer_copy['Age'] = customer_copy['Birthday'].apply(calculate_age)
```

```python
customer_copy.head()
```

```python
# Calculating the age range for analysis purpose

def agerange(age):
    if age < 10:
        return 'Under 10'
    elif age < 20:
        return '10-19'
    elif age < 30:
        return '20-29'
    elif age < 40:
        return '30-39'
    elif age < 50:
        return '40-49'
    elif age < 60:
        return '50-59'
    else:
        return '60 and above'

customer_copy['AgeRange'] = customer_copy['Age'].apply(agerange)
customer_copy.head()
```

### MERGING THE DATA'S

```python
sales_stores_product_customer_merged = pd.merge(left = sales_stores_product_merged,
                              right = customer,
                              left_on = 'CustomerKey',
                              right_on = 'CustomerKey' )

sales_stores_product_customer_merged.head()
```

```python
final_merged = pd.merge(left = sales_stores_product_customer_merged,
                              right = exrates,
                              left_on = ['CurrencyCode','OrderDate'],
                              right_on = ['Currency','Date'] 
                       )

final_merged
```
